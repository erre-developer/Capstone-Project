### Variables que se trabajaran para el modelo y dataframe

- Las variables que se trabajaran para el modelo seran las ["SalePrice"],["Lot.Area"],["Gr.Liv.Area"],["Heating.QC"],["Central.Air"]. En las variables de "Heating.QC" se clasificaran en 3 para generar variables Dummy, de igual forma para que "Central.Air" se trabajaran en Dummy para poder ingresarlas al modelo.

```python
df_analisis = df[["SalePrice","Lot.Area","Gr.Liv.Area","Heating.QC","Central.Air"]]
df_analisis
```

### Clasificación de la variable "Heating.QC" y "Central.Air"

De las 5 clasificaciones que tenia la variable "Heating.QC" (Ex,TA,Gd,Fa,Po) la trabajaremos para tenerlas en tres. De igual manera en "Central.Air" las pondremos en español para que se entiendan.

```python
df_analisis['Heating.QC'] = np.where( df['Heating.QC'] == "Ex", "Excelente" , 
                                       np.where( df['Heating.QC'] == "Gd", "Bueno" , "Tipico_pobre" ) )
```

```python
df_analisis['Central.Air'] = np.where( df['Central.Air'] == "Y", "SI" , "NO" )
```

```python
df_analisis
```

### Creración de nuevo Archivo Excel

Se crea un nuevo archivo de excel llamado ames_processed.xlsx, para el cual se trabajara en los modelos y analisis.
